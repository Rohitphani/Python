```python
students = [
    {"student_id": 1, "name": "Alice", "dep_id": 101},
    {"student_id": 2, "name": "Bob", "dep_id": 102},
    {"student_id": 3, "name": "Charlie", "dep_id": 101},
    {"student_id": 4, "name": "Diana", "dep_id": 103},
    {"student_id": 5, "name": "Bruce", "dep_id": 104}
]

student_df = spark.createDataFrame(students)
display(student_df)
```

```python
departments = [
    {"dep_id": 101, "dep_name": "Engineering"},
    {"dep_id": 102, "dep_name": "Science"},
    {"dep_id": 103, "dep_name": "Arts"},
    {"dep_id": 104, "dep_name": "Robotics"},
    {"dep_id": 105, "dep_name": "A.I"}
]

dep_df = spark.createDataFrame(departments)
display(dep_df)
```

```python
df = (
    student_df
    .join(dep_df, on = "dep_id", how = "inner")
)
display(df)  
```

```python
df = (
    student_df
    .join(dep_df, on = "dep_id", how = "left")
)
display(df)  
```

```python
df = (
    student_df
    .join(dep_df, on = "dep_id", how = "right")
)
display(df)  
```

```python
df = (
    student_df
    .join(
        dep_df,
        on = student_df["dep_id"] == dep_df["dep_id"],
        how = "right"
    )
)
display(df)  
```

```python
df = (
    df
    .join(
        dep_df,
        on = student_df["dep_id"] == dep_df["dep_id"],
        how = "left"
    )
    .filter(dep_df["dep_id"].isNull())
)
display(df)  
```
