```python
file_path = "/Volumes/workspace/default/amazon_datasets/amazon_sales_2025_INR_cleaned.csv"

amazon_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(file_path)
display(amazon_df)
```

```python
amazon_df = amazon_df.filter(F.col("Dimension") == "Cash on Delivery") 
display(amazon_df)
```

```python
display(
    amazon_df.select(
        F.col("Report_Section"),
        F.col("Dimension")
        )
)
```
