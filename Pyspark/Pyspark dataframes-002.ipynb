```python
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType, StringType, StructField, StructType
```

```python
file_path = "/Volumes/workspace/default/kaggle_datasets/bihar_election_results.csv"

bihar_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(file_path)
display(bihar_df)
```

```python
bihar_df.orderBy(F.col("Constituency Number").asc()).display()
```

```python
# groupby

participation_df = (
    bihar_df
    .groupBy("constituency name","constituency number")
    .count()
)

(
    participation_df
    .orderBy(F.col("constituency number").asc())
    .display()            
)
```

```python
# groupby another way multiple aggregations functions

participation_df = (
    bihar_df
    .groupBy("constituency name","constituency number")
    .agg(
        F.count("*").alias("Total participation"),
        F.min("Total Votes").alias("Minimum votes"),
        F.max("Total Votes").alias("Maximum votes"),
        F.avg("Total Votes").alias("Average votes")
    )
#         F.min("age").alias("Minimum age"),
#         F.max("age").alias("Maximum age"),
#         F.avg("age").alias("Average age")
# 
)
(
    participation_df
    .orderBy(F.col("constituency number").asc())
    .display()
)
```
