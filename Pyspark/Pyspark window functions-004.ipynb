```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window
# from pyspark.window import Window
```

```python
file_path = "/Volumes/workspace/default/emp_dataset"
# Emp_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(file path)
emp_df = spark.read.csv(file_path, header=True, inferSchema=True)
display(emp_df)
```

```python
emp_df_stats = (
    emp_df
    .groupBy("Gender")
    .agg(F.count("*").alias("Total Count"))
)
display(emp_df_stats)
```

```python
emp_df_stats = (
    emp_df
    .groupBy("City")
    .agg(
         F.sum(F.when(F.col("Gender") == "Male", 1).otherwise(0)).alias("Male"),
        F.sum(F.when(F.col("Gender") == "Female", 1).otherwise(0)).alias("Female")    
    )
)
display(emp_df_stats)
```
